## 1. Dispatches comparison

```python
# read the data
renewable_df = pd.read_csv(os.path.join(result_dir, "renewables_detail.csv"))
compare_renewable_df = pd.read_csv(os.path.join(comparison_result_dir, "renewables_detail.csv"))

# select the target gen
renewable_df = renewable_df.set_index("Generator").loc[renewable_gen]
compare_renewable_df = compare_renewable_df.set_index("Generator").loc[renewable_gen]

# add time index cols
renewable_df['Time Index'] = renewable_df.groupby("Generator").cumcount()
compare_renewable_df['Time Index'] = compare_renewable_df.groupby("Generator").cumcount()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
renewable_df.plot(x="Time Index", y="Output DA", ax=ax, label='Bidding DA Output')
compare_renewable_df.plot(x="Time Index", y="Output DA", ax=ax, label='DA Output')

ax.set_xlabel("Time [hr]", fontsize=LABEL_SIZE)
ax.set_ylabel("Power Output [MW]", fontsize=LABEL_SIZE)
ax.xaxis.set_minor_locator(MultipleLocator(5))

ax.tick_params(axis='both', which='major', labelsize=MAJOR_TICK_SIZE)
ax.tick_params(axis='both', which='minor', labelsize=MINOR_TICK_SIZE)

plt.rc('legend', fontsize = LEGEND_SIZE)

ax.grid()
ax.set_title(f"{renewable_gen} Day-ahead Power Output Comparison", fontsize=TITLE_SIZE)
```

## 3. Price comparison

```python
# read the data
renewable_bus_df = pd.read_csv(os.path.join(result_dir, "bus_detail.csv"))
compare_renewable_bus_df = pd.read_csv(os.path.join(comparison_result_dir, "bus_detail.csv"))

# select the target gen
bus_df = pd.read_csv(os.path.join(rts_dir, "bus.csv"))
renewable_bus_df = renewable_bus_df.merge(bus_df[["Bus ID", "Bus Name"]], left_on="Bus",right_on="Bus Name").set_index("Bus ID").loc[bus]
compare_renewable_bus_df = compare_renewable_bus_df.merge(bus_df[["Bus ID", "Bus Name"]], left_on="Bus",right_on="Bus Name").set_index("Bus ID").loc[bus]

# # add time index cols
renewable_bus_df['Time Index'] = renewable_bus_df.groupby("Bus ID").cumcount()
compare_renewable_bus_df['Time Index'] = compare_renewable_bus_df.groupby("Bus ID").cumcount()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
renewable_bus_df.plot(x="Time Index", y="LMP", ax=ax, label='Bidding RT LMP')
compare_renewable_bus_df.plot(x="Time Index", y="LMP", ax=ax, label='RT LMP')

ax.set_xlabel("Time [hr]", fontsize=LABEL_SIZE)
ax.set_ylabel("Locational Marginal Price [$/MW]", fontsize=LABEL_SIZE)
ax.xaxis.set_minor_locator(MultipleLocator(5))

ax.tick_params(axis='both', which='major', labelsize=MAJOR_TICK_SIZE)
ax.tick_params(axis='both', which='minor', labelsize=MINOR_TICK_SIZE)

plt.rc('legend', fontsize = LEGEND_SIZE)

ax.grid()
ax.set_title(f"Bus {bus} Real-time LMP Comparison", fontsize=TITLE_SIZE)
```

```python
renewable_df = renewable_df.reset_index().set_index("Time Index")
compare_renewable_df = compare_renewable_df.reset_index().set_index("Time Index")

renewable_df["LMP"] = renewable_bus_df["LMP"].values
renewable_df["LMP DA"] = renewable_bus_df["LMP DA"].values

compare_renewable_df["LMP"] = compare_renewable_bus_df["LMP"].values
compare_renewable_df["LMP DA"] = compare_renewable_bus_df["LMP DA"].values
```

```python
renewable_df["Revenue DA"] = renewable_df["LMP DA"] * renewable_df["Output DA"]
renewable_df["Revenue RT"] = renewable_df["LMP"] * (renewable_df["Output"] - renewable_df["Output DA"])

compare_renewable_df["Revenue DA"] = compare_renewable_df["LMP DA"] * compare_renewable_df["Output DA"]
compare_renewable_df["Revenue RT"] = compare_renewable_df["LMP"] * (compare_renewable_df["Output"] - compare_renewable_df["Output DA"])
```
